```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
data=pd.read_csv('Iris.csv')
```

```python
iris_setosa=data[data['Species']=='Iris-setosa']
```

```python
iris_setosa
```

```python
avg_sepallength=data['SepalLengthCm'].mean()
avg_sepalwidth=data['SepalWidthCm'].mean()
avg_petallength=data['PetalLengthCm'].mean()
avg_petalwidth=data['PetalWidthCm'].mean()
```

```python
plt.figure(figsize=(30, 10))
plt.plot(data['Id'], data['SepalLengthCm'], linestyle='-',s color='blue', linewidth=2, label='Sepal Length')
plt.axhline(y=avg_sepallength, color='r', linestyle='--', linewidth=3, label='Avg')
plt.xlabel('ID', fontsize=24)
plt.ylabel('Sepal Length', fontsize=24)
plt.title('Distribution of Sepal Length', fontsize=32)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(30, 10))
plt.plot(data['Id'], data['SepalWidthCm'], linestyle='-', color='blue', linewidth=2, label='Sepal Width')
plt.axhline(y=avg_sepalwidth, color='r', linestyle='--', linewidth=3, label='Avg')
plt.xlabel('ID', fontsize=24)
plt.ylabel('Sepal Width', fontsize=24)
plt.title('Distribution of Sepal Width', fontsize=32)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(30, 10))
plt.plot(data['Id'], data['PetalLengthCm'], linestyle='-', color='blue', linewidth=2, label='Petal Length')
plt.axhline(y=avg_petallength, color='r', linestyle='--', linewidth=3, label='Avg')
plt.xlabel('ID', fontsize=24)
plt.ylabel('Petal Length', fontsize=24)
plt.title('Distribution of Petal Length', fontsize=32)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(30, 10))
plt.plot(data['Id'], data['PetalWidthCm'], linestyle='-', color='blue', linewidth=2, label='Petal Width')
plt.axhline(y=avg_petalwidth, color='r', linestyle='--', linewidth=3, label='Avg')
plt.xlabel('ID', fontsize=24)
plt.ylabel('Petal Width', fontsize=24)
plt.title('Distribution of Petal Width', fontsize=32)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.legend()
plt.show()
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
plt.hist(data['SepalLengthCm'], bins=20, color='cyan', edgecolor='black', alpha=0.7)
plt.title('Histogram of Sepal Length', fontsize=16)
plt.xlabel('Volume', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(True)
plt.tight_layout()
plt.show()
```
