```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
data=pd.read_csv('thermodynamics.csv')
```

```python
avg_volume=data['Volume'].mean()
avg_temp=data['Temperature'].mean()
avg_pressure=data['Pressure'].mean()
avg_kinetic=data['Kinetic'].mean()
avg_potential=data['Potential'].mean()
avg_energy=data['Energy'].mean()
avg_density=data['Density'].mean()
```

```python
plt.figure(figsize=(30, 10))
plt.plot(data['Time'], data['Volume'], linestyle='-', color='blue', linewidth=2, label='Volume')
plt.axhline(y=avg_volume, color='r', linestyle='--', linewidth=3, label='Avg')
plt.xlabel('Time', fontsize=24)
plt.ylabel('Volume', fontsize=24)
plt.title('Time v Volume', fontsize=32)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(30, 10))
plt.plot(data['Time'], data['Temperature'], linestyle='-', color='blue', linewidth=2, label='Temperature')
plt.axhline(y=avg_temp, color='r', linestyle='--', linewidth=3, label='Avg')
plt.xlabel('Time', fontsize=24)
plt.ylabel('Temperature', fontsize=24)
plt.title('Time v Temperature', fontsize=32)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.legend()
plt.show()
```
